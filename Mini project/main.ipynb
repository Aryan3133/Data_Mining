```python
# Extract unique mappings
commodity_map = df[['Commodity_Code', 'Commodity_Description']].drop_duplicates().sort_values('Commodity_Code')
country_map = df[['Country_Code', 'Country_Name']].drop_duplicates().sort_values('Country_Code')
attribute_map = df[['Attribute_ID', 'Attribute_Description']].drop_duplicates().sort_values('Attribute_ID')
unit_map = df[['Unit_ID', 'Unit_Description']].drop_duplicates().sort_values('Unit_ID')

# Optional: Reset index for clean display
commodity_map.reset_index(drop=True, inplace=True)
country_map.reset_index(drop=True, inplace=True)
attribute_map.reset_index(drop=True, inplace=True)
unit_map.reset_index(drop=True, inplace=True)
```

```python
missing_country_code_rows = df[df['Country_Code'].isnull()]
print(missing_country_code_rows[['Country_Name', 'Country_Code']])
```

```python
# Group by Unit_Description, collect unique commodities as a list
grouped = df.groupby('Unit_Description')['Commodity_Description'].unique()

# Convert each list of commodities to a comma-separated string
grouped = grouped.apply(lambda x: ', '.join(x))

# Convert Series to a DataFrame with one row, units as columns
pivot_df = pd.DataFrame([grouped])

print(pivot_df)
```
