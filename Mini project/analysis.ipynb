```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

# Clean and prepare data
df = df.dropna(subset=['Value', 'Country_Name', 'Commodity_Description',
                       'Attribute_Description', 'Market_Year', 'Unit_Description'])
df['Market_Year'] = df['Market_Year'].astype(int)

# Create initial dropdowns
commodity_list = sorted(df['Commodity_Description'].unique())
commodity_dropdown = widgets.Dropdown(
    options=commodity_list,
    description='Commodity:',
    layout=widgets.Layout(width='45%')
)

attribute_dropdown = widgets.Dropdown(
    options=[],  # Will update dynamically
    description='Attribute:',
    layout=widgets.Layout(width='45%')
)

# Update attribute options based on selected commodity
def update_attribute_options(*args):
    selected_commodity = commodity_dropdown.value
    if selected_commodity:
        filtered_attrs = df[df['Commodity_Description'] == selected_commodity]['Attribute_Description'].unique()
        attribute_dropdown.options = sorted(filtered_attrs)

commodity_dropdown.observe(update_attribute_options, names='value')
update_attribute_options()  # initial fill

# Plotting function
def plot_line_chart(selected_commodity, selected_attribute):
    clear_output(wait=True)
    display(commodity_dropdown, attribute_dropdown)

    filtered_df = df[
        (df['Commodity_Description'] == selected_commodity) &
        (df['Attribute_Description'] == selected_attribute) &
        (df['Value'] > 0)
    ]

    if filtered_df.empty:
        print("No data available for this selection.")
        return

    # Calculate top 10 countries by total Value
    top_countries = (
        filtered_df.groupby('Country_Name')['Value']
        .sum()
        .sort_values(ascending=False)
        .head(10)
        .index
    )

    # Filter again for only top countries
    top_df = filtered_df[filtered_df['Country_Name'].isin(top_countries)]

    # Group and pivot
    grouped = top_df.groupby(['Market_Year', 'Country_Name'])['Value'].sum().reset_index()
    pivot_df = grouped.pivot(index='Market_Year', columns='Country_Name', values='Value').fillna(0)

    # Get unit(s)
    units = top_df['Unit_Description'].unique()
    unit_str = ', '.join(units)

    # Plot
    plt.figure(figsize=(12, 6))
    for country in pivot_df.columns:
        plt.plot(pivot_df.index, pivot_df[country], label=country)

    plt.title(f"{selected_attribute} of {selected_commodity} by Top 10 Producing Countries Over Years\nUnit(s): {unit_str}")
    plt.xlabel("Market Year")
    plt.ylabel(f"Value ({unit_str})")
    plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Interactive output
out = widgets.interactive_output(plot_line_chart, {
    'selected_commodity': commodity_dropdown,
    'selected_attribute': attribute_dropdown
})

# Display widgets and output
display(commodity_dropdown, attribute_dropdown, out)
```

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# Clean data
df = df.dropna(subset=['Value', 'Country_Name', 'Commodity_Description',
                       'Attribute_Description', 'Market_Year', 'Month'])
df['Market_Year'] = df['Market_Year'].astype(int)
df['Month'] = df['Month'].astype(int)

# Dropdown options
country_list = sorted(df['Country_Name'].unique())
commodity_list = sorted(df['Commodity_Description'].unique())
attribute_list = sorted(df['Attribute_Description'].unique())

country_dropdown = widgets.Dropdown(options=country_list, description='Country:', layout=widgets.Layout(width='40%'))
commodity_dropdown = widgets.Dropdown(options=commodity_list, description='Commodity:', layout=widgets.Layout(width='40%'))
attribute_dropdown = widgets.Dropdown(options=attribute_list, description='Attribute:', layout=widgets.Layout(width='40%'))

def regression_predict(selected_country, selected_commodity, selected_attribute):
    clear_output(wait=True)
    display(country_dropdown, commodity_dropdown, attribute_dropdown)

    # Filter data by selections
    data = df[(df['Country_Name'] == selected_country) &
              (df['Commodity_Description'] == selected_commodity) &
              (df['Attribute_Description'] == selected_attribute) &
              (df['Value'] > 0)]
    
    if data.empty:
        print("No data available for this selection.")
        return

    # Prepare features and target
    X = data[['Market_Year', 'Month']]
    y = data['Value']

    # One-hot encode categorical variables - for simplicity here we don't include categorical as features
    # because we already filter by them; but if needed, you can include more features.

    # Split train/test by year (e.g. last 20% years as test)
    years = sorted(data['Market_Year'].unique())
    if len(years) < 3:
        print("Not enough years of data for training.")
        return
    train_years = years[:int(len(years)*0.8)]
    test_years = years[int(len(years)*0.8):]

    train_mask = data['Market_Year'].isin(train_years)
    test_mask = data['Market_Year'].isin(test_years)

    X_train = X[train_mask]
    y_train = y[train_mask]
    X_test = X[test_mask]
    y_test = y[test_mask]

    # Model
    model = RandomForestRegressor(random_state=42, n_estimators=100)
    model.fit(X_train, y_train)

    # Predict test
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)

    # Predict next 5 years (using all months 1-12)
    last_year = max(years)
    future_years = np.repeat(np.arange(last_year+1, last_year+6), 12)
    future_months = np.tile(np.arange(1, 13), 5)
    future_X = pd.DataFrame({'Market_Year': future_years, 'Month': future_months})
    future_pred = model.predict(future_X)

    # Plot results
    plt.figure(figsize=(14,6))

    # Actual train + test
    plt.plot(data['Market_Year'].astype(str) + '-' + data['Month'].astype(str).str.zfill(2), data['Value'], label='Actual', marker='o')

    # Predicted test
    test_dates = X_test['Market_Year'].astype(str) + '-' + X_test['Month'].astype(str).str.zfill(2)
    plt.scatter(test_dates, y_pred, color='orange', label='Predicted (Test)', marker='x')

    # Future predictions
    future_dates = future_X['Market_Year'].astype(str) + '-' + future_X['Month'].astype(str).str.zfill(2)
    plt.plot(future_dates, future_pred, label='Predicted (Future)', linestyle='--')

    plt.xticks(rotation=90)
    plt.ylabel('Production Value')
    plt.title(f'Production Value Prediction for {selected_commodity} in {selected_country}\nAttribute: {selected_attribute}\nMAE on test set: {mae:.2f}')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Link widgets and function
out = widgets.interactive_output(regression_predict, {
    'selected_country': country_dropdown,
    'selected_commodity': commodity_dropdown,
    'selected_attribute': attribute_dropdown
})

display(country_dropdown, commodity_dropdown, attribute_dropdown, out)
```
